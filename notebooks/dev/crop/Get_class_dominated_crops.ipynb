### At this point, it is assumed that original images and masks have been exported from LabelBox to folders `masks` and `images`

```python
with open(".data/masks/classes.json", "r") as f:
    cls2idx = json.load(f)
```

```python
if not os.path.exists("./data/crops/asphalt"):
    os.makedirs("./data/crops/asphalt")

if not os.path.exists("./data/crops/concrete"):
    os.makedirs("./data/crops/concrete")
```

### Crop and save

```python
CROP_WIDTH = 300
CROP_HEIGHT = 300
THRESHOLD = 0.95
CROPS_PER_IMAGE = 10

def crop_from_one_image(image_file):
    try:
        print(image_file)
        mask_file = f"{image_file[:-3]}png"
        for surface in ["Asphalt", "Concrete"]:
            crops = None
            crops = get_class_dominated_crops(
                mask_file=os.path.join("./data/masks", mask_file),
                crop_size=(CROP_WIDTH, CROP_HEIGHT),
                dominating_class=cls2idx[surface],
                num_crop=CROPS_PER_IMAGE,
                threshold=THRESHOLD,
            )
            if crops is not None:
                img = np.array(Image.open(os.path.join("./data/images", image_file)))
                for i, crop in enumerate(crops):
                    img_crop = img[
                        crop[1] : (crop[1] + CROP_HEIGHT),
                        crop[0] : (crop[0] + CROP_WIDTH),
                        :,
                    ]
                    img_crop = Image.fromarray(img_crop)
                    img_crop.save(
                        os.path.join(
                            "./data/crops", surface.lower(), f"{image_file[:-4]}_crop_{i}.png"
                        )
                    )         
    except Exception as e:
            print(e)
```

```python
parallel(crop_from_one_image, os.listdir("./data/images/"))
```
