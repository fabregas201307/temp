```python
al_data = df_feature.copy(deep=True)
cols_forward_returns = ['returns_1mo_forward', 'returns_3mo_forward', 'returns_1yr_forward']
al_data[cols_forward_returns] = df_forward_returns[cols_forward_returns]
```
