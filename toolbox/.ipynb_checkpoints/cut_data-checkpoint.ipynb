```python
import pandas as pd
```

```python
df = pd.read_csv("Tommy_univ.csv")
```

```python
df
```

```python
df.columns
```

```python
cols_picked = ['date', 'sid', 'symbol', 'Current_Ratio', 'EPS_surprise',
              'Sector', 'price_to_book', 'Retained_Earnings', 'returns_1mo_forward', 'returns_3mo_forward']
```

```python
df_picked = df[cols_picked]
```

```python
cut_time = "2006-02-15"
```

```python
df_small = df_picked[df_picked['date']<cut_time]
```

```python
df_small
```

```python
columns_rename = dict()
columns_rename['Retained_Earnings'] = 'transactions'
columns_rename['Current_Ratio'] = 'Constant_nonConstant'
```

```python
columns_rename
```

```python
temp = df_small.rename(columns=columns_rename)
```

```python
temp.to_csv("Quantopian_data_picked_small.csv", index=None)
```
